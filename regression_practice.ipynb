```python
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import numpy as np
%matplotlib inline
```

```python
df = pd.read_csv('import85p.csv')
```

```python
df.head()
```

```python
# visualizing realtionship between features and target (part1)
sns.pairplot(df, x_vars = ['engine_size', 'hoursepower', 'city_mpg', 'highway_mpg'],
             y_vars = ['price'], height = 7, aspect = 0.7, kind = 'reg')
```

```python
# creating feature matrix and target column
feature_cols = ['engine_size', 'hoursepower', 'city_mpg', 'highway_mpg']
X = df[feature_cols]
y = df['price']

# creating linreg object
lm = LinearRegression()

from sklearn import metrics
```

```python
# splitting data set to train, test
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)

# fitting model
lm.fit(X_train, y_train)

# making predictions on test set
y_pred = lm.predict(X_test)

# calculating RMSE
print(np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
# range of price
print(y_test.min(), y_test.max())
```
