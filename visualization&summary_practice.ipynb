```python
import pandas as pd
```

```python
df = pd.read_csv("import85p.csv")
```

```python
df.head()
```

```python
# relationship between drive_wheel and price
import seaborn as sns
%matplotlib inline
sns.boxplot(x = "drive_wheels", y= "price", data = df)
```

```python
# relationship between price & engine size, hoursepower
sns.pairplot(df, x_vars = ['engine_size', 'hoursepower'], y_vars = ['price'], height = 7, aspect = 0.7, kind = 'reg')
```

```python
import scipy.stats as stats
# correlation coefficient for engine_size and price
pearson_coef, p_value = stats.pearsonr(df.engine_size, df.price)
```

```python
print(pearson_coef, p_value)
```
