```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import beta
from scipy import integrate
```

```python
# alpha, beta parameters
a, b = 2, 5

# probability vector
p = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
```

```python
# beta(2, 5) distribution
plt.plot(beta.pdf(p, a, b), color= "green")
plt.title("Beta(2,5)")
plt.xlabel("G")
plt.ylabel("Density")
plt.show()
```

```python
# Beta object
function = beta(a, b)

# lower and upper bound
l, u = 0.51, 1

# p(G > 50)
integrate.quad(function.pdf, l, u)
```

```python
# updated model
# alpha, beta parameters
a, b = a + 7, b + 17

# beta(9, 22) distribution
plt.plot(beta.pdf(p, a, b), color= "blue")
plt.title("Beta(9, 22)")
plt.xlabel("G")
plt.ylabel("Density")
plt.show()
```
