```python
from scipy.stats import beta
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
# prior parameters
a, b = 2, 5 

# "probability vector"
p = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
```

```python
# prior beta dist plot
fig, ax = plt.subplots()
ax.plot(beta.pdf(p, a, b), 'b-')
ax.set_title('PDF for probability of good grade')
ax.set_xlabel('probability of good grade')
ax.set_ylabel('density')
```

```python
# posterior parameters 
c, d = a + 3, b + 1

# posterior beta dist plot
fig1, ax1 = plt.subplots()
ax1.plot(beta.pdf(p, c, d), 'r-')
ax1.set_title('PDF for probability of good grade')
ax1.set_xlabel('probability of good grade')
ax1.set_ylabel('density')
```

```python
# plotting both dist together
figs, axs = plt.subplots()
axs.plot(beta.pdf(p, a, b), 'b-')
axs.plot(beta.pdf(p, c, d), 'r-')
axs.set_title('PDF for probability of good grade')
axs.set_xlabel('good grade')
axs.set_ylabel('density')
```

```python
import pandas as pd

# number of trials
n = 100000

prior = beta(a, b)
posterior = beta(c, d)

# draw 100k samples from prior dist
prior_samples = pd.Series([prior.rvs() \
                          for _ in range(n)])

# draw 100k samples from posterior dist
posterior_samples = pd.Series([posterior.rvs() \
                          for _ in range(n)])

# number of times posterior outperformed prior
post_win = sum(posterior_samples > prior_samples)

# result in percentage
post_win / n
```
