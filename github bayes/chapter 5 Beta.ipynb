```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
%matplotlib inline
```

```python
# alpha and beta parameters
a, b = 4, 6

# creating probability values vector
p = np.linspace(beta.ppf(0.01, a, b), beta.ppf(0.99, a, b), 100)
```

```python
# plotting distribution
fig, ax = plt.subplots()
ax.plot(beta.pdf(p, a, b), 'b-')
ax.set_title('Beta distribution for underlying bias of coin')
ax.set_xlabel('Bias')
ax.set_ylabel('Density')
```
